```python
import pandas as pd

df = pd.read_excel(r"G:\Datasets file\health_fitness_cleaned_dataset(AutoRecovered).xlsx")

df = df.loc[:, ~df.columns.str.contains("Unnamed")]

df["activity_cohort"] = pd.cut(
    df["daily_steps"],
    bins=[0, 5000, 10000, 20000],
    labels=["Low Activity", "Moderate Activity", "High Activity"]
)

df["sleep_cohort"] = pd.cut(
    df["hours_sleep"],
    bins=[0, 6, 8, 12],
    labels=["Poor Sleep", "Normal Sleep", "Good Sleep"]
)

df["stress_cohort"] = pd.cut(
    df["stress_level"],
    bins=[0, 3, 6, 10],
    labels=["Low Stress", "Moderate Stress", "High Stress"]
)

cohort_table = df.groupby(
    ["activity_cohort", "sleep_cohort", "stress_cohort"]
).size().reset_index(name="user_count")

print(cohort_table.head(20))

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x="activity_cohort", data=df)
plt.title("User Activity Cohorts")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x="sleep_cohort", data=df)
plt.title("User Sleep Cohorts")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(x="stress_cohort", data=df)
plt.title("User Stress Cohorts")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"G:\Datasets file\health_fitness_cleaned_dataset(AutoRecovered).xlsx")

df = df.loc[:, ~df.columns.str.contains("Unnamed")]

df["date"] = pd.to_datetime(df["date"])

df = df.sort_values(by="date")

daily_data = df.groupby("date")[["daily_steps","hours_sleep","stress_level"]].mean().reset_index()

daily_data["steps_rolling"] = daily_data["daily_steps"].rolling(window=7).mean()
daily_data["sleep_rolling"] = daily_data["hours_sleep"].rolling(window=7).mean()
daily_data["stress_rolling"] = daily_data["stress_level"].rolling(window=7).mean()

plt.figure(figsize=(8,4))
plt.plot(daily_data["date"], daily_data["daily_steps"])
plt.plot(daily_data["date"], daily_data["steps_rolling"])
plt.title("Steps Trend (with Rolling Average)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(daily_data["date"], daily_data["hours_sleep"])
plt.plot(daily_data["date"], daily_data["sleep_rolling"])
plt.title("Sleep Trend (with Rolling Average)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(daily_data["date"], daily_data["stress_level"])
plt.plot(daily_data["date"], daily_data["stress_rolling"])
plt.title("Stress Trend (with Rolling Average)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

df = pd.read_excel(r"G:\Datasets file\health_fitness_cleaned_dataset(AutoRecovered).xlsx")

df = df.loc[:, ~df.columns.str.contains("Unnamed")]

df = df.dropna(subset=["daily_steps","hours_sleep","stress_level","hydration_level","calories_burned","duration_minutes"])

df["daily_steps"] = df["daily_steps"].clip(0, 15000)
df["hours_sleep"] = df["hours_sleep"].clip(0, 8)
df["stress_level"] = df["stress_level"].clip(0, 10)
df["hydration_level"] = df["hydration_level"].clip(0, 4)

df["wellness_score"] = (
    (df["daily_steps"] / 15000) * 25 +
    (df["hours_sleep"] / 8) * 25 +
    (1 - (df["stress_level"] / 10)) * 25 +
    (df["hydration_level"] / 4) * 25
)

df["recovery_index"] = (
    (df["hours_sleep"] / 8) * 50 +
    (1 - (df["stress_level"] / 10)) * 50
)

df["activity_efficiency"] = df["calories_burned"] / df["duration_minutes"]

df["wellness_category"] = df["wellness_score"].apply(
    lambda x: "High" if x >= 75 else "Moderate" if x >= 50 else "Low"
)

df["recovery_category"] = df["recovery_index"].apply(
    lambda x: "Good" if x >= 70 else "Average" if x >= 40 else "Poor"
)

print(df[[
    "participant_id",
    "wellness_score",
    "wellness_category",
    "recovery_index",
    "recovery_category",
    "activity_efficiency"
]].head(10))

print("\nWellness Category Distribution:")
print(df["wellness_category"].value_counts())

print("\nRecovery Category Distribution:")
print(df["recovery_category"].value_counts())
```
